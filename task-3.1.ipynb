### *РЕШЕНИЕ*

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
```

```python
df_train ,df_test=train_test_split(df,random_state=42,test_size=0.33)
x=df_train.values
x=x[:,:-1]
y=df_test.values
y=y[:,:-1]
```

```python
ridge=Ridge(random_state=42)
model=ridge.fit(x,df_train.target)
```

```python
predicted=model.predict(y)
print(predicted, df_test.target)
```

```python
predictions_ground_truth_df=pd.DataFrame(list(zip(predicted,df_test.target)))
predictions_ground_truth_df.columns=['Predicted','Ground truth']
predictions_ground_truth_df.head()
```

```python
answer1=mean_squared_error(predicted, df_test.target)
print('mse 1: {0:.2f}'.format(answer1))
```
